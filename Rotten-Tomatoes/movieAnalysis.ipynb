```python
def createLists(df_pure, yr):
    date = []
    tomato = []
    aud = []
    
    breakDate = datetime(yr,4,15)
    
    for i,rowDate in enumerate(df_pure['in_theaters_date']):
        if(rowDate >= breakDate):
            date.append(rowDate.year)
            tomato.append(df_pure['tomatometer_avg_rating'].iloc[i])
            aud.append(df_pure['audience_top_critics_avg_rating'].iloc[i])
            
    return {'lists': [date, tomato, aud] }
```

```python
df_pure['in_theaters_date'] = pd.to_datetime(df_pure['in_theaters_date'], format = '%b %d, %Y')
```

```python
dataReturned = createLists(df_pure)
```

```python
results = dataReturned['lists']
```
